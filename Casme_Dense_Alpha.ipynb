```python
#creating pathfile for saving dense optical flow images
pathd=[]
for i in range(len(emotionp)):
    pathff=[]
    files=glob.glob(emotionp[i]+'*.jpg')
    for file in files:
        file=file.replace("\\", "/")
        #print(file)
        pathff.append(file)
    
    pathd.append(pathff)

#pathd[1]
```
